```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
# Ensure 'dataset3.csv' is in your working directory
ds = pd.read_csv("dataset3.csv")

# 2. Calculate the Correlation Matrix
# We filter for numeric columns only to avoid errors
corr_matrix = ds.select_dtypes(include=['number']).corr()

# 3. Isolate 'Pressure' and sort by strength
pressure_collinearity = corr_matrix['EXP_ADS (mL/g)'].sort_values(ascending=False)

print("--- Collinearity with 'EXP_ADS (mL/g)' ---")
print(pressure_collinearity)

# 4. Visual Check (Heatmap)
plt.figure(figsize=(6, 8))
sns.heatmap(
    corr_matrix[['EXP_ADS (mL/g)']].sort_values(by='EXP_ADS (mL/g)', ascending=False),
    annot=True, 
    cmap='coolwarm', 
    vmin=-1, 
    vmax=1
)
plt.title("Feature Correlation with Pressure")
plt.show()
```
